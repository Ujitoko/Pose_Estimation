```python
from xml.etree import ElementTree
import numpy as np
from PIL import Image

def xml_to_npy(src_path, filename, dst_dir):
    tree = ElementTree.parse(src_path)
    root = tree.getroot()

    root.keys()
    root.items()

    # rows, cols, dataを抽出
    rows = root.find(".//rows").text
    cols = root.find(".//cols").text
    data = root.find(".//data").text

    # rows, cols, dataを確認
    #print(rows)
    #print(cols)
    #print(data)

    img = data.split()
    img_np = np.array(img).reshape((int(rows), int(cols))).astype(float)
    img_np = np.multiply(img_np, 255)
    pil_img = Image.fromarray(np.uint8(img_np))
    pil_img.thumbnail((256, 256), Image.ANTIALIAS)
    
    pil_img.save(os.path.join(dst_dir, filename + ".jpg"), "JPEG", quality=100)
    #np.save(os.path.join(dst_dir, filename), img_np)

#x = "1_Sitting_pose_viewed_from_the_front_1/Training_data/Depth/366_test2_561_3501_0561.png.xml"
#xml_to_npy(x, " ")
```

```python
import os
def xmls_to_npy(src_dir, dst_dir):
    # dst_dirがなければ作る
    dst_dir = dst_dir
    if os.path.isdir(dst_dir) == False:
        os.mkdir(dst_dir)

    src_dir = src_dir
    filename_list_label = os.listdir(src_dir)
    for filename in filename_list_label:
        #print(filename)
        xml_to_npy(os.path.join(src_dir, filename), filename, dst_dir)

dst_dir = "1_Sitting_pose_viewed_from_the_front_1/Training_data/Depth_jpg"
src_dir = "1_Sitting_pose_viewed_from_the_front_1/Training_data/Depth"
xmls_to_npy(src_dir, dst_dir)
```

```python
import numpy as np
from PIL import Image
import os

def labelimg_to_npy(src_path, filename, dst_dir):
    img = Image.open(os.path.join(src_path))
    img = img.convert("RGB")
    img_np = np.array(img)
    #plt.imshow(np.array(img_np))
    
    num_to_color = {
        0: [255,123,0  ], # torso
        1: [0,  255,0  ], # head
        2: [0,  0,  255], # left upper arm
        3: [125,0,  125], # right upper arm
        4: [125,0,  0  ], # hip
        5: [245,0,  0  ], # left fore arm
        6: [0,  125,0  ], # right fore arm
        7: [0,  255,255], # left upper leg
        8: [255,0,  255], # right upper leg
        9: [255,255,0  ], # left lower leg
        10:[255,255,255], # right lower leg
        11:[0,  0,  0  ]  # background
    }
    
    print(img_np.shape)
    for row in range(img_np.shape[0]):
        for col in range(img_np.shape[1]):
            img_np[row][col] = num_to_color[img_np[row][col][0]]
    #plt.imshow(img_np)
    pil_img = Image.fromarray(np.uint8(img_np))
    pil_img.thumbnail((256, 256), Image.ANTIALIAS)
    pil_img.save(os.path.join(dst_dir, filename + ".jpg"), "JPEG", quality=100)

    #np.save(os.path.join(dst_dir, filename), img_np)

#x = "1_Sitting_pose_viewed_from_the_front_1/Training_data/Label"
#labelimg_to_npy(x, "366_test2_561_3501_0571.png", " ")
```
